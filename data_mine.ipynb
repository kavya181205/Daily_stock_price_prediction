```python
import pandas as pd
import yfinance as yf
```

```python
start_date="2020-06-09"
end_date="2025-11-29"

tata=yf.download(
    tickers="TATASTEEL.NS",
    interval="1d",
    auto_adjust=True,
    multi_level_index=False,
    start=start_date,
    end=end_date
)

reliance=yf.download(
    tickers="RELIANCE.NS",
    interval="1d",
    auto_adjust=True,
    multi_level_index=False,
    start=start_date,
    end=end_date
)

mrf=yf.download(
    tickers="MRF.NS",
    interval="1d",
    auto_adjust=True,
    multi_level_index=False,
    start=start_date,
    end=end_date
)
```

```python
tata_old = pd.read_csv("TATASTEEL.NS.csv")
reliance_old = pd.read_csv("RELIANCE.NS.CSV")
mrf_old = pd.read_csv("MRF.NS.CSV")
tata_old.info()
```

```python
tata_old.drop(columns=['Adj Close'],inplace=True)
reliance_old.drop(columns=['Adj Close'],inplace=True)
mrf_old.drop(columns=['Adj Close'],inplace=True)
```

```python
tata_old.info()
```

```python
tata.reset_index(inplace=True)
reliance.reset_index(inplace=True)
mrf.reset_index(inplace=True)
tata.info()
```

```python
tata_new = pd.concat([tata_old,tata])
reliance_new = pd.concat([reliance_old,reliance])
mrf_new = pd.concat([mrf_old,mrf])
```

```python
tata_new['company'] = 'tata'
reliance_new['company'] = 'reliance'
mrf_new['company'] = 'mrf'
```

```python
df = pd.concat([tata_new,reliance_new,mrf_new])
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values(by='Date',inplace=True)
df.head()
```

```python
df.to_csv("final_destination.csv")
```
